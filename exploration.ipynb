```python
import tabula
import pandas as pd
```

```python
dfs = tabula.read_pdf('./permuta.pdf')
df_aux = dfs[0]
if len(dfs)>1:
    for df in dfs:
        df_aux = pd.concat(df_aux, df)
df= df_aux
    
```

```python
df.columns = ['gp_wk', 'wk', 'gp_cg', 'cg', 'mat', 'road_map']
df['gp_wk'] = df['gp_wk'].apply(lambda x: x.split(' ')[1])
df['mat'] = df['road_map'].apply(lambda x: x.split(' ')[0])
df['road_map'] = df['road_map'].apply(lambda x: x.split(' ')[1])
df
```
